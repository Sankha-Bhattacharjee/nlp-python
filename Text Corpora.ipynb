```python
from nltk.corpus import genesis
```

```python
# Various text collections available under genesis text corpus are viewed by fileids method.
genesis.fileids()
```

```python
genesis.sents('english-kjv.txt')
```

```python
for fileid in genesis.fileids():
    n_chars = len(genesis.raw(fileid))
    n_words = len(genesis.words(fileid))
    n_sents = len(genesis.sents(fileid))
    print("****",fileid, ":")
    print("characters length: ", n_chars)
    print("words length: ", n_words)
    print(f"sentence length: ", n_sents)
    print(f"Average word length, Average words length:: ({int(n_chars/n_words)},  {int(n_words/n_sents)})")
```

```python
from nltk import ConditionalFreqDist
from nltk.corpus import brown
```

```python
brown.words(categories='adventure')
```

```python
cfd = ConditionalFreqDist([ 
(genre, word) 
for genre in brown.categories() 
for word in brown.words(categories=genre) ])
cfd
```

```python
cfd.conditions()
```

```python
cfd['adventure']['.']
```

```python
cfd['government']['leadership']
```

```python
 cfd.tabulate(conditions=['government', 'humor', 'reviews'], samples=['leadership', 'worship', 'hardship'])
```

```python
cfd.tabulate(conditions=['government', 'humor', 'reviews'], samples=['leadership', 'worship', 'hardship'], cumulative = True)
```

```python
from nltk.corpus import names
for fid in names.fileids():
    for name in names.words(fid):
        print(fid, name)
```

```python
from nltk.corpus import names
nt = [(fid.split('.')[0], name[-1])    for fid in names.fileids() 
  for name in names.words(fid) ]
```

```python
nt
```

```python
cfd2 = ConditionalFreqDist(nt)
cfd2.conditions()
```

```python
cfd2['male']
```

```python
sum([cfd2['male'][x] for x in cfd2['male']])
```

```python
#The expression checks if the last characters in females occur more frequently than the last characters in males.
sum([cfd2['female'][x] for x in cfd2['female']]) > sum([cfd2['male'][x] for x in cfd2['male']])
```

```python
cfd2.tabulate(samples=['a', 'e'])
```
